```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,4],"C":[15,16,17,18]})
pd.merge(var1 ,var2,on="A")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[15,16,17,18]})
pd.merge(var1 ,var2,how="inner", indicator=True) 
#in how we can give inner, left, right, outer
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,5],"B":[15,16,17,18]})
pd.merge(var1 ,var2,right_index=True, left_index=True) 
```

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,12,13,14])
pd.concat([sr1,sr2],axis=1)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2],"B":[15,16]})
pd.concat([var1,var2],axis=1)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2],"B":[15,16]})
pd.concat([var1,var2],axis=1,join="inner", keys=["d1","d2"])
```
